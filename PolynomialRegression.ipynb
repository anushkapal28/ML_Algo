```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('Ice_cream selling data.csv')
```

```python
df.head()
```

```python
X = df[['Temperature (°C)']].values
y = df["Ice Cream Sales (units)"].values
```

```python
lr = LinearRegression()
lr.fit(X,y)
y_pred = lr.predict(X)
```

```python
r2_linear = r2_score(y,y_pred)
print("R2 score for Linear Regression : ", r2_linear)
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Linear fit')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Converts X into a polynomial feature matrix.
#degree=2 means the model will use X² as an additional feature.
poly = PolynomialFeatures(degree = 2)
 # Transforms X into a matrix containing [1, X, X²] (a constant, X, and X²).
X_poly = poly.fit_transform(X)
# Initialize linear regression model
poly_lr = LinearRegression()
# Train the model using polynomial features
poly_lr.fit(X_poly, y)
 # Make predictions using polynomial regression
y_pred_poly = poly_lr.predict(X_poly)
```

```python
r2_poly = r2_score(y, y_pred_poly)
print(f'R² score for polynomial regression: {r2_poly:.4f}')
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_poly, color='green', label='Polynomial fit')
plt.title('Polynomial Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
poly = PolynomialFeatures(degree = 15)
X_poly = poly.fit_transform(X)
poly_lr = LinearRegression()
poly_lr.fit(X_poly, y)
y_pred_poly = poly_lr.predict(X_poly)
```

```python
r2_poly = r2_score(y, y_pred_poly)
print(f'R² score for polynomial regression: {r2_poly:.4f}')
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_poly, color='green', label='Polynomial fit')
plt.title('Polynomial Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
